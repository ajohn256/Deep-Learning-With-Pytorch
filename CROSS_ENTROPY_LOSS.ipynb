```python
import torch
import numpy as np
import torch.nn as nn
```

Cross-entropy loss, also known as logarithmic loss or log loss, is a popular loss function used in machine learning to measure the performance of a classification model. It quantifies the difference between the predicted probability distribution and the true distribution of the data

cross-entropy-loss-for-multiclass-classification.png

```python
def CrossEntropy(actual,predicted):
  loss = -np.sum(actual * np.log(predicted))
  return loss
```

```python
y = np.array([1,2,5])
y_pred_good = np.array([0.8,1.2,4.5])
y_pred_bad = np.array([2.5,0.3,0.1])
```

```python
l1 = CrossEntropy(y,y_pred_good)
l2 = CrossEntropy(y,y_pred_bad)
```

```python
print(f"Loss1 : {l1:.3f}")
print(f"Loss2 : {l2:.3f}")
```

# **WITH PYTORCH**

```python
loss = nn.CrossEntropyLoss()

y = torch.tensor([0])
y_pred_good = torch.tensor([[2.0,1.0,0.1]])
y_pred_bad = torch.tensor([[0.5,2.0,0.3]])
```

```python
l1 = loss(y_pred_good,y)
l2 = loss(y_pred_bad,y)
```

```python
print(f"Loss1 : {l1.item():.3f}")
print(f"Loss2 : {l2.item():.3f}")
```

```python
_, predictions1 = torch.max(y_pred_good,1)
_, predictions2 = torch.max(y_pred_bad,1)
```

```python
print(predictions1)
print(predictions2)
```
