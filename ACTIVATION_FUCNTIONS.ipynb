```python
import torch
import torch.nn as nn
import numpy as np
```

# **Common activation functions include**


*   Softmax
*   TanH
*   ReLU
*   Leaky ReLU
*   Sigmoid
*   Step function

```python
class NeuralNetwork(nn.Module):
  def __init_(self,input_size,output_size):
    super(NeuralNetwork,self).__init__()
    self.layer1=nn.Linear(input_size,output_size)
    self.relu = nn.ReLU()
    self.layer2 = nn.Linear(output_size,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self,x):
    out = self.layer1(x)
    out = self.relu(out)
    out = self.layer2(x)
    out = self.sigmoid(out)
    return out
```

# **Alternatively**

```python
class NeuralNetwork(nn.Module):
  def __init_(self,input_size,output_size):
    super(NeuralNetwork,self).__init__()
    self.layer1=nn.Linear(input_size,output_size)
    self.layer2 = nn.Linear(output_size,1)

  def forward(self,x):
    out = nn.ReLU(self.layer1(x))
    out = nn.Sigmoid(self.layer2(x))
    return out
```
